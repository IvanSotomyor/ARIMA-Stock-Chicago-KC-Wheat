#Librerias

```python
#Librerias
import pandas as pd
import yfinance as yf
```

#Data Frame

```python
# Definir el DF
ticker = 'ZW=F'
start_date = '2019-01-01'
end_date = '2022-10-10'

df = yf.download(ticker, start_date)
```

```python
#Imprimir df
print(df)
```

#Preprocesamiento

```python
#Cambiar tipo de fecha
df.index = pd.to_datetime(df.index).strftime('%Y-%m-%d')
```

```python
#Trecio en dolares de las toneladas
df ['Precio_Ton'] = df ['Close'] * 0.367437
df.head()
```
